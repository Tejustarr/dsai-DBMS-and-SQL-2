# Session 4 Lab
Demonstrating joins in SQLite.

```python
import sqlite3, pandas as pd
conn = sqlite3.connect('session4_lab.db')
print('Connected')
```

```python
# Inner join employees and departments
pd.read_sql_query('SELECT e.first_name, d.dept_name FROM employees e JOIN departments d ON e.dept_id = d.dept_id;', conn)
```

```python
# Left join
pd.read_sql_query('SELECT e.first_name, d.dept_name FROM employees e LEFT JOIN departments d ON e.dept_id = d.dept_id;', conn)
```

```python
# Cross join
pd.read_sql_query('SELECT e.first_name, d.dept_name FROM employees e CROSS JOIN departments d;', conn)
```
