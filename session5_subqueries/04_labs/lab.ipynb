Demonstrating subqueries in SQLite.

```python
import sqlite3, pandas as pd
conn = sqlite3.connect('session5_lab.db')
print('Connected')
```

```python
# Employees with salary > average salary
pd.read_sql_query('SELECT * FROM employees WHERE salary > (SELECT AVG(salary) FROM employees);', conn)
```

```python
# Departments with no employees
pd.read_sql_query('SELECT dept_name FROM departments d WHERE NOT EXISTS (SELECT 1 FROM employees e WHERE e.dept_id=d.dept_id);', conn)
```
