Demonstrating window functions and constraints.

```python
import sqlite3, pandas as pd
conn = sqlite3.connect('session6_lab.db')
print('Connected')
```

```python
# Rank employees by salary
pd.read_sql_query('SELECT emp_id, salary, RANK() OVER (ORDER BY salary DESC) as rnk FROM employees;', conn)
```

```python
# Recursive CTE example (may not work in old SQLite)
pd.read_sql_query('WITH RECURSIVE nums(n) AS (SELECT 1 UNION ALL SELECT n+1 FROM nums WHERE n<5) SELECT * FROM nums;', conn)
```
